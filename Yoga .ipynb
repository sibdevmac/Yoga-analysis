<b> 3. Are yoga classes required for every type to learn every pose? 

```python
difficulty_counts = df["Difficulty"].value_counts()
print(difficulty_counts)
```

```python
def class_requirement(level):
    if level == "Beginner":
        return "No"
    elif level == "Intermediate":
        return "Optional"
    else:
        return "Yes"
df["Class_Required"] = df["Difficulty"].apply(class_requirement)
print(df[["Pose", "Difficulty", "Class_Required"]])
```

```python
class_counts = df["Class_Required"].value_counts()
print(class_counts)
```

```python
class_counts.plot(kind='bar')
plt.title("Do Yoga Classes Required?")
plt.xlabel("Class Requirement")
plt.ylabel("Number of Poses")
plt.show()
```

```python
type_class = pd.crosstab(df["Type"], df["Class_Required"])
print(type_class)
```

```python
sns.heatmap(type_class, annot=True, fmt='d')
plt.title("Yoga Type vs Class Requirement")
plt.show()
```

<b> 4. Is there any correlation between body parts and types of yoga, and if it really removes the pain? 

```python
correlation_table = pd.crosstab(df["Body_Part"], df["Type"])
sns.heatmap(correlation_table, annot=True, fmt='d')
plt.title("Body Part vs Yoga Type")
plt.show()
```

```python
pain_df = df[df["Health_Condition"].str.contains("Pain", case=False)]
print(pain_df[["Pose", "Body_Part", "Type"]])
```

<b> 5. What are the best Yoga combinations for improving my standing, sitting and back pain?  

```python
standing = df[df["Type"] == "Standing"]["Pose"]
print("Standing Improvement:\n", standing.tolist())

sitting = df[df["Type"].isin(["Seated", "Meditation"])]["Pose"]
print("\nSitting Improvement:\n", sitting.tolist())

back_pain = df[df["Health_Condition"].str.contains("Back Pain")]["Pose"]
print("\nBack Pain Relief:\n", back_pain.tolist())
```

<b> 6. Finally, predict which yoga could cure which diseases. 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["Health_Condition"])
y = df["Pose"]
model = DecisionTreeClassifier()
model.fit(X, y)
def predict_yoga(disease):
    X_test = vectorizer.transform([disease])
    return model.predict(X_test)[0]
```

```python
print(predict_yoga("Stress"))
```

```python
print(predict_yoga("Fatigue"))
```

```python
print(predict_yoga("Obesity"))
```

```python
print(predict_yoga("Hypertension"))
```

```python
print(predict_yoga("Back Pain"))
```

<b> 7. Map the body parts as per Yoga pose with a human shape.

```python
import matplotlib.image as mpimg
img = mpimg.imread("human body.png.png") 
fig, ax = plt.subplots(figsize=(6,10))
ax.imshow(img)
ax.axis('off')
```

```python
h, w, _ = img.shape
print(h, w)
```

```python
body_coordinates = {
    "Mind": (0.5*w, 0.08*h),     
    "Chest": (0.5*w, 0.28*h),    
    "Spine": (0.5*w, 0.45*h),    
    "Core": (0.5*w, 0.58*h),     
    "Legs": (0.5*w, 0.80*h)      
}
```

```python
import matplotlib.patches as mpatches
```

```python
body_map = df.groupby("Body_Part")["Pose"].apply(list)
for part, (x, y) in body_coordinates.items():
    poses = body_map.get(part, [])
    label = f"{part}\n{len(poses)} poses"
    ax.text(x, y, label, color='red', fontsize=10, ha='center')
```

```python
fig, ax = plt.subplots(figsize=(6,10))
ax.imshow(img)
ax.axis('off')
for part, (x, y) in body_coordinates.items():
    poses = df[df["Body_Part"] == part]["Pose"].tolist()
    ax.scatter(x, y, s=len(poses)*400, alpha=0.4)
    label = f"{part}\n{poses[0]}\n+{len(poses)-1} more"
    ax.text(x, y, label, fontsize=9, ha='center',
            bbox=dict(facecolor='white', alpha=0.6))
legend_elements = [
    mpatches.Patch(color='blue', label='Mind'),
    mpatches.Patch(color='orange', label='Chest'),
    mpatches.Patch(color='green', label='Spine'),
    mpatches.Patch(color='red', label='Core'),
    mpatches.Patch(color='purple', label='Legs')
]
ax.legend(handles=legend_elements, loc='upper right')
plt.title("Yoga Impact + Pose Examples")
plt.show()
```
